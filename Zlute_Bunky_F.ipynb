Zaměstnanec vyjižděl soubory z databáze obsahujicí data x krát. Když provedl vyjetí v následujicím období, tak se v novým souboru vyjela data z předchozícho vyjetí návíc. Řekneme že první soubor obsahuje data $a_1$ až $a_n$, druhý soubor nešlo vyjet jinak než s daty z prvního souboru navíc
takže $a_1$ až $a_{n+5}$, třetí soubor obdobně $a_1$ až $a_{n+10}$. Když už byly k dispozici 3 soubory, určitá data byla potřeba úpravit. Každý soubor opravoval jiný zaměstnanec.

- Soubor s datama $a_1$ až $a_5$ (první soubor) opravovál Marek.

- Nasledujicí den soubor s datama $a_6$ až $a_{10}$ (druhý soubor) který zaroveň obsahuje data $a_1$ až $a_5$ (z prvního souboru) opravoval Fanda. Fanda si mohl všimnout dat z prvního souboru který Marek zapomněl upravit. Tak to opravil za něj v druhém souboru, jelikož je to jediný soubor který má k dispozici.

- Třetí a posledni soubor který obsahuje nová data a zároveň data z prvního a druhého souboru se skladá z $a_1$ až $a_{15}$ opravovál Tonda. Tonda si taky mohl všimnout data z prvního a druhého souboru který zapomněli Marek a Fanda upravit. Tonda to opravil za ně v posledním souboru.


Poté co změny byly provedený v každým souboru, zaměstnavatel si přál mít všechna data v jednom souboru aby měl čerstvý přehled všeho (data která neměla byt upravena a zaroveň ta upravena). 

Data ma svoje ID který jednodušše vyznačuje pořadí. V některých souborech byla přidaná data navíc jelikož si zaměstnanci všimli že tam chybí. Přidana data nemáji přiřazený žádný ID (jelikož žádný zaměstnanec nevěděl jestli následujicí den bude někdo opravovat další soubor a proto nevěděli jaký je poslední ID položky).

V tomhle chaosu je ale jediná vyhoda. Kdekoliv v souboru se šáhlo tak se automatický změnila barvá buňky na např. žlutou, aby bylo jasné co je jinák než v originálu. Data nebylo možne upravovát manualně jelikož se jednalo o 40 000 řádku dat.

```python
from styleframe import StyleFrame, utils
import pandas as pd
import numpy as np
import openpyxl
import os
from os import walk
```

<h1>Načtení dat</h1>
<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Zadefinuju cestu do složky kde jsou soubory a jejich počet.
</font>
<br>

```python
cesta = r'Input'
soubory = next(os.walk(cesta))[2]
poc_soub = len(soubory)
```

<h1>Zpracování dat</h1>
<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    sframe_list: množina obsahujicí tabulky ve StyleFrame
    <br>
    pd_list: množina obsahujicí tabulky ve Pandas
</font>

```python
sframe_list = []
pd_list = []
for i in range(poc_soub):
    cesta_soubor = cesta + "\\" + soubory[i]
    s = StyleFrame.read_excel(cesta_soubor, read_style = True, use_openpyxl_styles = False) #Načtení ve Styleframe
    p = pd.read_excel(cesta_soubor) #Načtení v Pandas
    sframe_list.append(s) #Přidat do listu
    pd_list.append(p) #Přidat do listu
```

```python
print(sframe_list[0])
```

```python
pd_list[0]
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Funkce která nahradí žlutý buňky znakama ('@@@@@@'), jinak zapiše nulu
</font>

```python
def yellow_cells(cell):
    if cell.style.bg_color in {'FFFFFF00'}:
        return "@@@@@@"
    else:
        return 0
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Aplikovat na množinu načtenou ve styleframe (sframe_list). Zároveň to automatický překonvertuje tabulky do typu pandas 
    dataframe
</font>

```python
zluty_list = []
for i in range(len(sframe_list)):
    b = sframe_list[i].applymap(yellow_cells)
    zluty_list.append(b)
```

```python
zluty_list[0]
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Funkce 
        <font color=blue>
            zmeneny_radek
        </font>
    : Jestli jsou všechny řádky nulový napíše I, jinak O

```python
def zmeneny_radek(radek):
    if ((radek['ID'] == 0) & (radek['A'] == 0) & (radek['B'] == 0) & (radek['C'] == 0) & (radek['D'] ==0)):
        hodnota = "I"
    else:
        hodnota = "O"
    return hodnota
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Aplikujeme jí na množinu kde máme symboly místo žlutych buněk (zluty_list).
</font>

```python
for i in range(len(zluty_list)):
    zluty_list[i]['Zmeneny radek'] = zluty_list[i].apply(zmeneny_radek, axis = 1)
```

```python
zluty_list[0]
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Funkce 
        <font color=blue>
            najit_indexy_navic
        </font>
    : iterovat přes řádky sloupce "ID" aby se zjistilo jestli je nějakej "ID" rovny hodnotě <b>np.nan</b> nebo <b>0</b> 
    (Když někdo přidal nový řádek, nechal buňku prázdnou, nebo případně napsal nulu. V pandas se to mohlo načíst jako np.nan 
    nebo 0). Jestli nějaky takový řádek existuje, přida se jeho pořadí do množiny.

```python
def najit_indexy_navic(df):
    b = 0
    mnozina = []
    for j in df:
        navic_ind = []
        for i in range(len(j['ID'])):
            if ((j['ID'][i] == 0) | (pd.isna(j['ID'][i]) == True)):
                navic_ind.append(i)    
        mnozina.append(navic_ind)
    return mnozina
```

```python
radky_navic = najit_indexy_navic(pd_list)
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Je pouze jeden řádek navíc a to ve třetím souboru.
</font>

```python
radky_navic
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Odebrat řádek navíc z množiny kde jsou nahrazené žluté buňky
</font>

```python
radky_pridane = []
for i in range(len(radky_navic)):
    for j in range(len(radky_navic[i])):
        radky_pridane.append(pd_list[i].loc[radky_navic[i][j]]) #přídam řádky do množiny
        zluty_list[i] = zluty_list[i].drop([radky_navic[i][j]]) #odstraním řádky
        pd_list[i] = pd_list[i].drop([radky_navic[i][j]]) #odstraním řádky
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Uložit zvlášť poslední sloupec ve kterém je aplikovaná funkce 
    <font color=blue>
        zmeneny_radek
    </font>

```python
zluty_sloupec = []
for i in range(len(zluty_list)):
    j = zluty_list[i].iloc[:,[-1]].copy() #poslední sloupec
    zluty_sloupec.append(j) #množina s poslednimi sloupci
```

```python
zluty_sloupec[0]
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Přídat poslední sloupec tabulce načtenou v Pandas, najít upravený řádky, odstranít neupravený
</font>

```python
zlute_zmeny = []
for i in range(len(zluty_sloupec)):
    k = pd.concat([pd_list[i], zluty_sloupec[i]], axis = 1) #spojit pandas tabulku a posledni sloupec
    zlute_zmeny.append(k) #přidat do množiny
    zlute_zmeny[i] = zlute_zmeny[i].loc[zlute_zmeny[i]['Zmeneny radek'] == 'O'] #vybrat pouze upravený řádky
    zlute_zmeny[i] = zlute_zmeny[i].iloc[:,:-1].copy() #odstranít poslední sloupec
```

```python
zlute_zmeny[0]
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Spojením všech tabulek v množině <b>zlute_zmeny</b> máme celkové upravený řádky
</font>

```python
zmeny_spojene = pd.concat(zlute_zmeny)
```

```python
zmeny_spojene
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Stalo se mi když v upravených řádcích byla prázdná buňka (NaN) a update-oval jsem Archiv se upravenýma řádkama tak hodnoty
    NaN ignorovalo. Proto je potřeba hodnoty NaN nahradit nějakým speciálním symbolem a na konci nahradit zase zpátky.
</font>

```python
zmeny_spojene = zmeny_spojene.replace(np.nan, '@@n@@', regex = True)
```

```python
zmeny_spojene
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Načtení Archivu
</font>

```python
archiv = pd.read_excel(r'Input_Archiv\Archivos.xlsx')
```

```python
archiv
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Nastavit sloupec "ID" jako index a update-ovat Archiv, upravenými řádky
</font>

```python
ar = archiv.copy()
az = zmeny_spojene.copy()

ar.set_index('ID', inplace = True)
az.set_index('ID', inplace = True)

ar.update(az)

vysl = ar.reset_index()
```

```python
vysl
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Nahradit zpátky symboly s NaN
</font>

```python
vysl = vysl.replace('@@n@@', np.nan, regex = True)
```

```python
vysl
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Konvertovát řádky do typu DataFrame
</font>

```python
radkyy = pd.DataFrame(radky_pridane)
```

```python
radkyy = radkyy.reset_index(drop = True)
```

```python
radkyy
```

<br>
<font style="font-family:Times New Roman;font-size:17px;color:#4d0099;">
    Přidané řádky musí mít přidělén ID vetší (nebo v tomto připádě menší) než má poslední položka v Archivu. Proto jsem 
    vygeneroval "ID" sloupec který má "ID" čísla, podle toho kolik je přidaných řádku. Pak jsem "ID" čísla přidál přislušným
    řádkum.
</font>

```python
posledni_index = archiv['ID'].iloc[-1]
df1 = pd.DataFrame([[posledni_index-i] for i in range(1,(len(radkyy)+1))], columns=['ID'])
```

```python
df1
```

```python
radkyy[['ID']] = radkyy[['ID']].fillna(df1)
```
